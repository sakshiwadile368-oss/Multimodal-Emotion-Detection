```python
from collections import Counter

# Example predictions from your three models
face_emotion = "happy"
text_emotion = "joy"
speech_emotion = "happy"

# Normalize labels
emotion_map = {
    "joy": "happy",
    "happiness": "happy",
    "sadness": "sad",
    "anger": "angry",
    "fearful": "fear",
    "surprised": "surprise",
    "neutral": "neutral"
}

face_emotion = emotion_map.get(face_emotion.lower(), face_emotion.lower())
text_emotion = emotion_map.get(text_emotion.lower(), text_emotion.lower())
speech_emotion = emotion_map.get(speech_emotion.lower(), speech_emotion.lower())

# Majority Voting
votes = [face_emotion, text_emotion, speech_emotion]

final_emotion = Counter(votes).most_common(1)[0][0]

print("Face Emotion   :", face_emotion)
print("Text Emotion   :", text_emotion)
print("Speech Emotion :", speech_emotion)

print("\nFinal Emotion :", final_emotion)
```

```python
recommendations = {
    "happy": "😊 Keep smiling! Enjoy your day.",
    "sad": "❤️ Talk to someone you trust and take some rest.",
    "angry": "😌 Take deep breaths and relax.",
    "fear": "🌸 Stay calm. Everything will be okay.",
    "surprise": "🎉 Hope it is a pleasant surprise!",
    "neutral": "🙂 Have a productive day."
}

print("\nRecommendation:")
print(recommendations.get(final_emotion, "No recommendation available."))
```
