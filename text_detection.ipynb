```python
from transformers import pipeline

classifier = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    top_k=1
)

text = input("Enter Text : ")

result = classifier(text)

emotion = result[0][0]['label']

print("Detected Emotion :", emotion)
```
